# Functions

The basic sintax of a function in C++ is as follows:

```cpp
return_type FunctionName(parameter_list) {
  // Body of function here.
}
```

However, if we don't want to return anything, we can use `void`:
```cpp
void FunctionName(parameter_list) {
  // Body of function here.
}
```

````c++
#include <iostream>
#include <vector>

using std::cout;
````

````c++
int AdditionFunction(std::vector<int> vec) {
    int total_sum = 0;
    
    for (int i=0; i < vec.size(); i++) {
        total_sum += vec[i];
    }
    
    return total_sum;
}
````

````c++
std::vector<int> v {1, 2, 3};
cout << AdditionFunction(v) << "\n";
````
